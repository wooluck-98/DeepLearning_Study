```python
seoul_data = commercial[commercial['시'] == '서울특별시']
seoul_data.tail()
```

```python
city_type = set(seoul_data['시'])
city_type, len(city_type)
```

```python
seoul_chicken_data = seoul_data[seoul_data['상권업종소분류명'] == '후라이드/양념치킨']
set(seoul_chicken_data['상권업종소분류명'])
```

```python
groupdata = seoul_chicken_data.groupby("구")
group_by_category = groupdata['상권업종소분류명']
chicken_count_gu = group_by_category.count()
sorted_chicken_count_gu = chicken_count_gu.sort_values(ascending=False)
sorted_chicken_count_gu
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = "Malgun Gothic"
```

```python
plt.figure(figsize=(10, 5))
plt.bar(sorted_chicken_count_gu.index, sorted_chicken_count_gu)
plt.title("구에 따른 치킨가게 수의 합계")
plt.xticks(rotation = 45)
plt.show()
```

```python
# Q1. 4월 유동인구가 가장 많은 구는 어디인가요?
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = "Malgun Gothic"

data = pd.read_csv("data/population04.csv")
groupby_population = data.groupby("군구")['유동인구수'].sum()
sorted_population = groupby_population.sort_values(ascending = True)
plt.figure(figsize=(10, 5))
plt.bar(sorted_population.index, sorted_population)
plt.title("2020년 4월 서울 군구별 유동인구 수")
plt.xlabel("군구")
plt.ylabel("유동인구 수 (명)")
plt.xticks(rotation=-45)
plt.show
```

```python
# Q2. 4월과 7월의 강남구의 일별 유동인구는 어떤가요?
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = "Malgun Gothic"

month_4 = pd.read_csv("data/population04.csv")
month_7 = pd.read_csv("data/population07.csv")
gangnam_population_4 = month_4[month_4['군구'] == "강남구"]
gangnam_population_7 = month_7[month_7['군구'] == "강남구"]
groupby_date_4 = gangnam_population_4.groupby("일자")['유동인구수'].sum()
groupby_date_7 = gangnam_population_7.groupby("일자")['유동인구수'].sum()

x_label_4 = []
x_label_7 = []
for date in groupby_date_4.index:
    x_label_4.append(str(date))
for date in groupby_date_7.index:
    x_label_7.append(str(date))
plt.figure(figsize=(15, 5))
plt.title("4월과 7월 일자별 강남구 유동인구수")
plt.xlabel("날짜")
plt.ylabel("유동인구 수(천만명)")
plt.plot(x_label_4, groupby_date_4)
plt.plot(x_label_7, groupby_date_7)
plt.xticks(rotation=-90)
plt.show()
```
