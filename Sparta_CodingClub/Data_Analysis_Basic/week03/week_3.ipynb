```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
text = open("data/Sequence_01.txt", encoding = "utf-8")
text = text.read()
text = text.replace("\n", " ")
text
```

```python
result = ""
for number in range(1, 15):
    index = "{:02}".format(number)
    filename = "Sequence_" + index + ".txt"
    print(filename)
    text = open("data/"+filename, 'r', encoding = "utf-8")
    result += text.read().replace("\n", " ")
result
```

```python
import re
pattern = "[^\w\s]"
text = re.sub(pattern=pattern, repl="", string=result)
text
```

```python
import matplotlib.font_manager as fm
for f in fm.fontManager.ttflist:
    if "Gothic" in f.name:
        print(f.fname)
```

```python
font_path = "C:\Windows\Fonts\malgunsl.ttf"
wc = WordCloud(font_path = font_path, background_color = "white")
wc.generate(text)
plt.figure(figsize=(50, 50))
plt.axis("off")
plt.imshow(wc)
plt.show()
```

```python
mask = np.array(Image.open('./data/sparta.png'))
wc = WordCloud(font_path=font_path, background_color="white", mask=mask)
wc.generate(text)


f = plt.figure(figsize=(50,50))
f.add_subplot(1,2,1)
plt.imshow(mask, cmap=plt.cm.gray)
plt.title("Original Stencil", size=40)
plt.axis("off")
f.add_subplot(1,2,2)
plt.imshow(wc, interpolation='bilinear')
plt.title('Twitter Generated Cloud', size=40)
plt.axis("off")
plt.show()
f.savefig('./wordcloud_ex1.png')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = "Malgun Gothic"
```

```python
sparta_data = pd.read_csv('data/enrolleds_detail.csv')
sparta_data
```

```python
format = "%Y-%m-%dT%H:%M:%S.%f"
sparta_data['done_date_time'] = pd.to_datetime(sparta_data['done_date'], format=format)
sparta_data
```

```python
sparta_data["done_date_time_weekday"] = sparta_data['done_date_time'].dt.day_name()
sparta_data
```

```python
weekdata = sparta_data.groupby("done_date_time_weekday")['user_id'].count()
```

```python
weeks = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekdata = weekdata.agg(weeks)
weekdata
```

```python
plt.figure(figsize=(10,5))
plt.bar(weekdata.index, weekdata)
plt.title('요일별 수강 완료 수강생 수')
plt.xlabel('요일')
plt.ylabel('수강생(명)')
plt.xticks(rotation=90)
plt.show()
```

```python
sparta_data['done_date_time_hour'] = sparta_data['done_date_time'].dt.hour
hour_data = sparta_data.groupby("done_date_time_hour")['user_id'].count()
hour_data = hour_data.sort_index()
hour_data
```

```python
plt.figure(figsize=(10, 5))
plt.plot(hour_data.index, hour_data)
plt.title("시간별 수강 완료 사용자 수")
plt.xlabel("시간")
plt.ylabel("사용자(명)")
plt.xticks(np.arange(24))
plt.show()
```

```python
sparta_data_pivot = pd.pivot_table(sparta_data, values='user_id', aggfunc='count',
              index = ["done_date_time_weekday"],
              columns = ["done_date_time_hour"]).agg(weeks)
sparta_data_pivot
```

```python
plt.figure(figsize=(14, 5))
plt.pcolor(sparta_data_pivot)
plt.title("요일뵬 종료 시간 히트맵")
plt.xticks(np.arange(0.5, len(sparta_data_pivot.columns), 1), sparta_data_pivot.columns)
plt.yticks(np.arange(0.5, len(sparta_data_pivot.index), 1), sparta_data_pivot.index)
plt.xlabel("시간")
plt.ylabel("요일")
plt.colorbar()
plt.show()
```

```python
# Q1. 최애곡 가사 워드 클라우드 만들기
import numpy as np
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = open("data/trip.txt", 'r', encoding="utf-8")
result = text.read().replace("\n", " ")

for _ in range(10):
    result += result

mask = np.array(Image.open("data/insta.jpg"))
font_path = "C:\Windows\Fonts\malgunsl.ttf"
wc = WordCloud(font_path = font_path, background_color="white", mask = mask)
wc.generate(result)

f = plt.figure(figsize=(50, 50))
plt.axis("off")
plt.imshow(wc)
```

```python
# Q2. 월요일과 화요일의 수강완료 시간 비교
import pandas as pd
import matplotlib.pyplot as plt

sparta_data_monday = sparta_data[sparta_data['done_date_time_weekday'] == "Monday"]
sparta_data_monday = sparta_data_monday.groupby("done_date_time_hour").count()['user_id']

sparta_data_tuesday = sparta_data[sparta_data['done_date_time_weekday'] == "Tuesday"]
sparta_data_tuesday = sparta_data_tuesday.groupby("done_date_time_hour").count()['user_id']

plt.figure(figsize=(10, 5))
plt.plot(sparta_data_monday.index, sparta_data_monday)
plt.plot(sparta_data_tuesday.index, sparta_data_tuesday)
plt.title("시간별 수강 완료자 수")
plt.xticks(np.arange(24))
plt.show()
```
